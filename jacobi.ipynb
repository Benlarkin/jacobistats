```
# imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# this allows plots to appear directly in the notebook
%matplotlib inline
```

```
# read data into a DataFrame
data = pd.read_csv('https://raw.githubusercontent.com/Benlarkin/jacobistats/master/stats.txt')
data.head()
threadcount = data.values.tolist()
# add missing data which is misaligned for some reason
numthreads = [1]
elapsed = [1067.489597]
for t in threadcount:
  s = t[0].split()
  numthreads.append(int(s[0]))
  elapsed.append(float(s[1]))

#return % difference of values in decimal form
def dif(a, b):
  return -((a-b)/((a+b)/2)) * 100

# calculate percent difference per increased thread
percent = [None]
for num in range(1, len(elapsed)):
  percent.append(dif(elapsed[num], elapsed[num-1] ))
```

```

xi = list(range(len(numthreads)))

fig, ax1 = plt.subplots()
color = 'tab:red'
ax1.set_xlabel('NOTH')
ax1.set_ylabel('elapsed time(s)', color=color)
ax1.plot(xi, elapsed, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('speed increase %', color=color)
ax2.plot(xi, percent, color=color)
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_xlim(xmin=0)

fig.tight_layout()
plt.xticks(xi, numthreads)

plt.show()
```
